```python
import re
from google.colab import drive
drive.mount('/content/gdrive/',force_remount=True)
```

```python
data=open('/content/gdrive/MyDrive/myposts.txt',"r")
data_content=data.read()
```

```python
output = re.sub(r'#[\w-]+', '', data_content)
```

```python
output
```

```python
text_file = open("/content/gdrive/MyDrive/myposts.txt", "w")
text_file.write(output)
text_file.close()
```
